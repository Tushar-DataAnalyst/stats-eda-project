```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Stats\stats_data.csv")
```

```python
print("Mean",df['Order_Value'].mean())
print ("Mode", df['Order_Value'].mode())
print ("Mean",df['Order_Value'].median())
print("Max", df ['Order_Value'].max())
print("Min", df ['Order_Value'].min())
print("Skew", df['Order_Value'].skew())
print("std", df['Order_Value'].std())

Q1 = df['Order_Value'].quantile(0.25)
Q3 = df['Order_Value'].quantile(0.75)

IQR = Q3-Q1

lower = Q1 -IQR*1.5
upper = Q3+IQR*1.5

#df[(df['Order_Value'] < lower) | (df['Order_Value'] > upper)]

df['Outlier'] = ((df['Order_Value'] < lower) | (df['Order_Value'] > upper)).astype(int)
print(df['Outlier'])
print(df[['Customer_ID', 'Order_Value', 'Outlier']])

cc_return = df.groupby('Payment_Method')['Return_Flag'].mean() * 100
print(cc_return.round(2))



```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid", palette="dark")
sns.histplot(df['Order_Value'], bins=6, kde=True, color='skyblue')
plt.axvline(df['Order_Value'].mean(),color = "yellow" , linestyle = "--", linewidth = 2, label= f'Mean: Rs.{df["Order_Value"].mean():,.0f}')
plt.axvline(df['Order_Value'].median(), color='grey', linestyle=':', linewidth=2, label=f'Median: Rs.{df["Order_Value"].median():,.0f}')
plt.title('Order_Vlaue Histogram',fontsize= 15, fontweight ='bold')
plt.xlabel('Order Value Rs.')
plt.ylabel('frequency')
plt.legend()
plt.show()


plt.figure(figsize= (7,8))
sns.boxplot(x='Segment', y=df['Order_Value'],palette='pastel', width = 0.5, hue ='Return_Flag', data=df)
plt.title('Boxplot', fontsize = 10, fontweight ='bold')
plt.xlabel('Segment')
plt.ylabel('Order_value Rs.')

plt.show()

plt.figure(figsize = (8,6))
plt.bar(cc_return.index, cc_return.values, color= ['red','green','green'], edgecolor ='white')
plt.title('Return Status', fontsize =10, fontweight ='bold')
plt.show()

```

```python
corr = df[['Age', 
           'Monthly_Orders', 
           'Order_Value', 
           'Customer_Tenure', 
           'Support_Calls', 
           'Satisfaction_Score']].corr()

print(corr.round(2))
```

```python
plt.figure(figsize = (8,5))
sns.heatmap(corr, 
            annot=True,      
            cmap='coolwarm', 
            fmt='.2f')
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
